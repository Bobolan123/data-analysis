```python
import pandas as pd

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data"
df = pd.read_csv(url, header=None)
df
```

```python
column_name = ['symboling', 'normalise', 'make', 'fuel', 'engine type', 'doors', 'body style', 'drive-wheel', 'engine location']
for i in range(9, 26):
    column_name.append(f'col_{i}')
df.columns = column_name
df
```

```python
mean = df[df['normalise'] !='?']['normalise'].astype('int').mean()
df['normalise'].replace('?', mean, inplace=True)
df.normalise = df['normalise'].astype('int')
df
```

```python
set(df.doors)
```

```python
df.doors.replace('five', 1, inplace=True)
```

```python
df['drive-wheel'] = df['drive-wheel'].replace('4wd','four wheel drive')
set(df['drive-wheel'])
```
