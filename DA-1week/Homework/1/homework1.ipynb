```python
# pip install xlrd lib
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_excel('Sample - Superstore.xls')
df.info()
df
```

```python
df[['Category', 'Profit']].groupby('Category').sum('Profit').plot(kind="barh", xlabel="Profit of each category")

df[['Sub-Category', 'Profit']].groupby('Sub-Category').sum('Profit').plot(kind="barh", xlabel="Profit sub-category")
```

```python
df_state_city_profit = df[['State', 'City', 'Profit']]

df_state_city_profit[df_state_city_profit['Profit'] < 0].groupby('State').sum('Profit').sort_values('Profit', ascending=False).plot(kind='barh')
plt.tight_layout()
plt.show()

```

```python
df.isnull().sum()
```

```python
df[df['Postal Code'].isnull()] = df['Postal Code'].mean()

df['Postal Code'].isnull().sum()
```

```python
plt.hist(df['Profit'], bins=20)
plt.xlabel('Profit')
plt.ylabel('Frequency')
plt.title('Profit Distribution')
plt.show()
```

```python
profit_group = df.groupby(['Product ID', 'Product Name'])['Profit'].sum().reset_index()

profit_group = profit_group.sort_values('Profit', ascending=False)
profit_group = profit_group.head(10)
profit_group = profit_group.drop(0)

profit_group
```

```python
fig, ax = plt.subplots()
bar_colors = ['tab:red', 'tab:blue', 'tab:red', 'tab:orange', 'tab:green', 'tab:purple', 'tab:brown', 'tab:pink', 'tab:gray']

ax.barh(profit_group['Product Name'], profit_group['Profit'], color=bar_colors)

plt.show()
```
